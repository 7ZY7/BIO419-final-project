```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
df=pd.read_excel('Data.xlsx')
```

```python
df
```

```python
df_Male=df.loc[df['Victim Sex']=='Male']
df_male=df_Male.groupby(['State']).count()
df_male1=df_male.drop('Year',1)

df_Fem=df.loc[df['Victim Sex']=='Female']
df_female=df_Fem.groupby(['State']).count()
df_female1=df_female.drop('Year',1)
df_F_M1 =[df_male1,df_female1] 
df_F_M=pd.concat(df_F_M1,axis=1)
df_F_M.plot(kind='bar',color=['blue','pink'],stacked=True)
plt.legend(['Male','Female'])

plt.title('Victim Sex and incidence ')
```

```python
df_Fem['State'].value_counts().plot.pie(fontsize=7,autopct='%.2f')
plt.title('Female victim')
```

```python
df_Male['State'].value_counts().plot.pie(autopct='%.2f')
plt.title('Male Victim')
```

```python
df2=pd.read_excel('Data2.xlsx')
df2
```

```python
df_Jan=df2.loc[df2['Month']=='January']
df_Jan1=df_Jan.groupby(['State']).sum()
```

```python
df_Feb=df2.loc[df2['Month']=='February']
df_Feb2=df_Feb.groupby(['State']).sum()
```

```python
df_March=df2.loc[df2['Month']=='March']
df_Mar3=df_March.groupby(['State']).sum()

df_Apr=df2.loc[df2['Month']=='April']
df_Apr4=df_Apr.groupby(['State']).sum()

df_May=df2.loc[df2['Month']=='May']
df_May5=df_May.groupby(['State']).sum()

df_June=df2.loc[df2['Month']=='June']
df_June6=df_June.groupby(['State']).sum()

df_July=df2.loc[df2['Month']=='July']
df_July7=df_July.groupby(['State']).sum()

df_Aug=df2.loc[df2['Month']=='August']
df_Aug8=df_Aug.groupby(['State']).sum()

df_Sep=df2.loc[df2['Month']=='September']
df_Sep9=df_Sep.groupby(['State']).sum()

df_Oct=df2.loc[df2['Month']=='October']
df_Oct10=df_Oct.groupby(['State']).sum()

df_Nov=df2.loc[df2['Month']=='November']
df_Nov11=df_Nov.groupby(['State']).sum()

df_Dec=df2.loc[df2['Month']=='December']
df_Dec12=df_Dec.groupby(['State']).sum()
```

```python
df_Month =[df_Jan1,df_Feb2,df_Mar3,df_Apr4,df_May5,df_June6,df_July7,df_Aug8,df_Sep9,df_Oct10,df_Nov11,df_Dec12] 
```

```python
df_AMonth=pd.concat(df_Month,axis=1)


df_AMonth.T.plot.box()
plt.title('Variation of incident rate in 12-month by states ')
```
